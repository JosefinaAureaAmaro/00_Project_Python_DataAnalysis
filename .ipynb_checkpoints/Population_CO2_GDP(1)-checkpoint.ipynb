```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
```

# Population and GDP dataset

```python
file = "population_AQ_Full.csv"
population_file = pd.read_csv(file,encoding = "ISO-8859-1")
population_file
```

```python
population_file_1 = population_file.rename(index=str, columns={"Country Name_population": "Country_Name"})
population_file_1.head()
```

# CO2 dataset

```python
csv_path = "../../homework/Original CO2 emissions dataset.csv"
CO2data_df = pd.read_csv(csv_path, encoding = 'unicode_escape')
CO2data_df.head()
```

```python
CO2data_df_headers=list(CO2data_df.columns.values)
CO2data_df_years = CO2data_df_headers[-59:]
CO2data_df_years_drop = CO2data_df_years[1:51]
CO2data_df_1= CO2data_df.drop(CO2data_df_years_drop,axis=1)
CO2data_df_1.head()
```

```python
CO2data_df_2 = CO2data_df_1.drop('2010', axis = 1)
CO2data_df_2.head()
```

# Mergetable 

```python
merged_df = CO2data_df_2.merge(population_file_1, on= 'Country_Name')
merged_df.head()
```

```python
merged_df_1 = merged_df.rename(columns = {'2011':'2011_CO2','2012':'2012_CO2', '2013':'2013_CO2','2014':'2014_CO2','2015':'2015_CO2','2016':'2016_CO2','2017':'2017_CO2'})
merged_df_1.head()
```

```python
merged_df_2 = merged_df_1.nlargest(50, '2017_population')
merged_df_2.head()
```

```python
corr = merged_df_2.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
merged_df_2.dropna(inplace=True)
```

```python
plt.figure(figsize=(20,10))

y_axis=list(merged_df_2['2017_population'])
x_axis = np.asarray(list(merged_df_2['2017_CO2']))

slope, intercept, r_value, p_value, std_err = stats.linregress(x_axis,y_axis)
line = slope * x_axis + intercept


plt.scatter(x_axis, y_axis, marker="x", facecolors="black", edgecolors="black", alpha=0.75)
plt.plot(x_axis, line)
plt.grid()
plt.title("CO2 Emission (Mean Exposure) around the world in 2017")
plt.xlabel("CO2 Emission (Mean Exposure)")
plt.ylabel("Population(Millions)")

plt.savefig("../Project_1_AirPollution_vs_GDPperCapita/population vs CO2 2017.png")
plt.show()
```

```python
plt.figure(figsize=(20,10))

y_axis=list(merged_df_2['2017_population'])
x_axis = np.asarray(list(merged_df_2['2017_CO2']))

slope, intercept, r_value, p_value, std_err = stats.linregress(x_axis,y_axis)
line = slope * x_axis + intercept

plt.scatter(x_axis, y_axis, marker="x", facecolors="black", edgecolors="black", alpha=0.75)
plt.plot(x_axis, line)
plt.xlim(0, 750)
plt.grid()

plt.title("CO2 Emission (Mean Exposure) around the world in 2017")
plt.xlabel("CO2 Emission (Mean Exposure")
plt.ylabel("Population(Millions)")

plt.savefig("../Project_1_AirPollution_vs_GDPperCapita/population vs CO2 2017_exclude outliers.png")
plt.show()
```

```python
plt.figure(figsize=(20,10))
y_axis=list(merged_df_2['2017_gdp'])
x_axis = np.asarray(list(merged_df_2['2017_CO2']))

# Generated linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(x_axis,y_axis)
# slope = float(slope)
line = slope * x_axis + intercept

plt.scatter(x_axis, y_axis, marker="o", facecolors="blue", edgecolors="blue", alpha=0.75)
plt.plot(x_axis, line)
plt.grid()
plt.title("CO2 Emission (Mean Exposure) around the world in 2017")
plt.xlabel("CO2 Emission (Mean Exposure")
plt.ylabel("GDP")

plt.savefig("../Project_1_AirPollution_vs_GDPperCapita/GDP vs CO2 2017.png")
plt.show()
```
