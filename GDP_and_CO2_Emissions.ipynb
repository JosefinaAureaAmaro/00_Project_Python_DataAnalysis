```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

```python
csv_path = "../Project_1_AirPollution_vs_GDPperCapita/Original CO2 emissions dataset.csv"
CO2data_df = pd.read_csv(csv_path, encoding = 'unicode_escape')
CO2data_df.head()
```

```python
CO2data_df_headers=list(CO2data_df.columns.values)
CO2data_df_years = CO2data_df_headers[-59:]
CO2data_df_years_drop = CO2data_df_years[1:51]
CO2data_df_1= CO2data_df.drop(CO2data_df_years_drop,axis=1)
CO2data_df_1.head()
```

```python
CO2_pct_change = CO2data_df_1.pct_change(axis='columns')
```

```python
csv_path = "../Project_1_AirPollution_vs_GDPperCapita/GDPbyCountry.csv"
GDPbyCountry_df = pd.read_csv(csv_path, encoding = 'unicode_escape')
GDPbyCountry_df.head()
```

```python
GDPbyCountry_df_headers=list(GDPbyCountry_df.columns.values)
GDPbyCountry_df_years_drop_2018 = list([GDPbyCountry_df_headers[-1]])
GDPbyCountry_df_1= GDPbyCountry_df.drop(GDPbyCountry_df_years_drop_2018,axis=1)
GDPbyCountry_df_1.head()
```

```python
GDPCO2_df = pd.merge(CO2data_df_1, GDPbyCountry_df_1, on="Country")
GDPCO2_df.head()
```

```python
GDPCO2_df = GDPCO2_df.rename(columns={"2010_x": "2010_CO2_Emissions", "2011_x": "2011_CO2_Emissions", "2012_x": "2012_CO2_Emissions",
                                      "2013_x": "2013_CO2_Emissions", "2014_x": "2014_CO2_Emissions", "2015_x": "2015_CO2_Emissions", "2016_x": "2016_CO2_Emissions", "2017_x": "2017_CO2_Emissions"})

GDPCO2_df = GDPCO2_df.rename(columns={"2010_y": "2010_GDP", "2011_y": "2011_GDP", "2012_y": "2012_GDP",
                                      "2013_y": "2013_GDP", "2014_y": "2014_GDP", "2015_y": "2015_GDP", "2016_y": "2016_GDP", "2017_y": "2017_GDP"})

GDPCO2_df.head()
```

```python
GDPCO2_df.to_csv('GDP CO2 Emissions.csv', sep=',' , encoding= 'utf-8', index=False)
```
