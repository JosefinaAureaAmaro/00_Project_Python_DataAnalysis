```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data = "GDPbyCountry.csv"


# Read the City and Ride Data
city_data_df = pd.read_csv(city_data, encoding="iso-8859-1")

city_data_df = city_data_df.drop(['2010', '2011', '2012', '2018'], axis=1)

city_data_df['2013'] = city_data_df['2013'].replace('no data', np.nan)
city_data_df['2017'] = city_data_df['2017'].replace('no data', np.nan)
city_data_df['2016'] = city_data_df['2016'].replace('no data', np.nan)
city_data_df['2015'] = city_data_df['2015'].replace('no data', np.nan)
city_data_df['2014'] = city_data_df['2014'].replace('no data', np.nan)

city_data_df["2013"]= city_data_df["2013"].astype(float)
city_data_df["2017"]= city_data_df["2017"].astype(float)
city_data_df["2016"]= city_data_df["2016"].astype(float)
city_data_df["2015"]= city_data_df["2015"].astype(float)
city_data_df["2014"]= city_data_df["2014"].astype(float)


city_data_df = city_data_df.nlargest(15, '2017')

# # Display the data table for preview
city_data_df = city_data_df.reset_index(drop=True)

city_data_df = city_data_df.sort_values(by='2017', ascending=True)
city_data_df.head(15)
```

```python
city_data_df.dtypes
```

```python
y_axis = np.arange(len(city_data_df['Country']))
bars_in_cities = city_data_df['2017']

plt.barh(y_axis, bars_in_cities, color="b", align="center")
tick_locations = [value for value in y_axis]
plt.yticks(tick_locations, city_data_df['Country'])

plt.title("GDP by Country")
plt.xlabel("GDP ($M)")
```
