```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

```python
csv_path = "../Project_1_AirPollution_vs_GDPperCapita/PM25Data.csv"
PM25data_df = pd.read_csv(csv_path, encoding = 'unicode_escape')
PM25data_df.head()
```

```python
PM25data_df_1 = PM25data_df.drop(['Country Code', 'Indicator Name', 'Indicator Code'],axis=1)
PM25data_df_1.head()
```

```python
PM25data_df_2 = PM25data_df_1.sort_values(by=['2017'],ascending=False)
PM25data_df_2 = PM25data_df_2.loc[:,:'2017'].round(0)
PM25data_df_2.head(10)
```

```python
Country = PM25data_df_2.set_index('Country Name')
df_plot_by_country = Country.T
df_plot_by_country
```

```python
df_plot_by_country.index
```

```python
list_of_countries = list(df_plot_by_country.columns)
```

```python
nepal = list(df_plot_by_country['Nepal'])
niger = list(df_plot_by_country['Niger'])
qatar = list(df_plot_by_country['Qatar'])
india = list(df_plot_by_country['India'])
saudiarabia = list(df_plot_by_country['Saudi Arabia'])
egypt = list(df_plot_by_country['Egypt, Arab Rep.'])
southasiaida = list(df_plot_by_country['South Asia (IDA & IBRD)'])
southasia = list(df_plot_by_country['South Asia'])
    
plt.figure(figsize=(20,10))

plt.fill_between(df_plot_by_country.index, nepal)
plt.fill_between(df_plot_by_country.index, niger)
plt.fill_between(df_plot_by_country.index, qatar)
plt.fill_between(df_plot_by_country.index, india)
plt.fill_between(df_plot_by_country.index, saudiarabia)
plt.fill_between(df_plot_by_country.index, egypt)
plt.fill_between(df_plot_by_country.index, southasiaida)
plt.fill_between(df_plot_by_country.index, southasia)

plt.title("PM25 Air Pollution by Top 10 Countries")
plt.xlabel("Year")
plt.ylabel("PM2.5 Air Pollution")

plt.legend(['Nepal', 'Niger', 'Qatar', 'India', 'Saudi Arabia', 'Egypt, Arab Rep.', 'South Asia (IDA & IBRD)', 'South Asia'])

plt.savefig("../Project_1_AirPollution_vs_GDPperCapita/PM25_Air_Pollution_by_Countries.png")

plt.show()
```
